```python
import pandas as pd

def get_csv_df():
    return pd.read_csv('titanic.csv', index_col="PassengerId")
```

## GroupBy

```python
get_csv_df().groupby("Age").mean()
```

```python
get_csv_df().groupby("Age").count()
```

```python
get_csv_df().groupby("Age").value_counts()
```

## Sorting

```python
sorted = get_csv_df().sort_values(by="Age")
print(type(sorted))
sorted
```

## Concat

```python
df = get_csv_df()
df1 = df.head()
df2 = df.tail()

pd.concat([df1, df2])
```

## Column Values

```python
df = get_csv_df()

print(type(df['Name'].values))
df['Name'].values.tolist()
```

## Iterating Rows

```python
df = get_csv_df()
for r in df.iterrows():
    print(r)
```
