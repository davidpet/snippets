I was told that making new decorators was now considered bad practice but that it's ok to use ones that are already defined, such as in Angular. I'm not sure of the current status.


```typescript
(() => {
  // Decorator function
  function log(target: any, key: string, descriptor: PropertyDescriptor) {
    const originalMethod = descriptor.value;

    descriptor.value = function (...args: any[]) {
      console.log(
        `Method ${key} called with arguments: ${JSON.stringify(args)}`
      );
      const result = originalMethod.apply(this, args);
      console.log(`Method ${key} returned: ${JSON.stringify(result)}`);
      return result;
    };

    return descriptor;
  }

  // Class with a decorated method
  class MyClass {
    @log
    greet(name: string): string {
      return `Hello, ${name}!`;
    }
  }

  // Create an instance of MyClass
  const myInstance = new MyClass();

  // Call the decorated method
  const result = myInstance.greet("John");
  console.log(result);
})();
```
