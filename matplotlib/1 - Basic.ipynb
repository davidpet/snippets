## Import the library

pyplot subpackage gives a *stateful matlab-like* interface.

```python
import matplotlib.pyplot as plt
from IPython import display

## %matplotlib inline           # a lot of people use this but it seems like it's already the default anyway
```

## Create a dataset

```python
x = [1, 2, 3, 4, 5]
y = [e**2 for e in x]

print(x)
print(y)
```

## Create a figure with 1 subplot (1 row, 1 col)
These 3 lines must be in the *same cell* because the first line make the physical graph.

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## Again with a more sophisticated dataset

This is the example from the quickstart tutorial.

```python
import numpy as np

# create evenly spaced x values from 0 to 2*pi.
x = np.linspace(0, 2 * np.pi, 200)
# Get the sin values for all those points.
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## plt.show() not actually needed for Jupyter notebooks

```python
fig, ax = plt.subplots()
ax.plot(x, y)
pass         # This is just here to make the last line not return anything (because that is shown before plot).
```

## Implicit API

You can pretend *plt* is the last figure/axes created and also that you have one by default if you start using it.

```python
plt.plot(x, y)
```

## Export to file

```python
fig, ax = plt.subplots()
ax.plot(x, y)

fig.savefig('data/figure.png')  # The folder must exist or you get an ERROR!
```

## Re-exporting to same file

```python
fig, ax = plt.subplots()
ax.plot(y, x)  # Switch order so we can tell the difference

fig.savefig('data/figure.png')  # File is OVERWRITTEN!
```

## Redrawing in a loop

```python
print('this text will be gone')

for i in range(10):
    display.clear_output(wait=True)     # Clear all output text and graphs
    plt.plot(x, y)                      # Do a new plot
    plt.show()                          # For some reason the graph clearing doesn't take effect until this
```

## Leaving Out X Values

This is key as it allows you to plot *any sequence* of values without having to generate x values for them.

```python
plt.plot(y)      # x is assumed to be 0-based index of y array
```
