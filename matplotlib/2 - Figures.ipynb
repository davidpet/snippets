```python
import matplotlib.pyplot as plt
```

## Empty Figure

You could add axes later.


```python
fig0 = plt.figure()  # an empty figure with no Axes
```

## 1 Axes Figure

ax1 is an _axes_, containing 2 _axis objects_.


```python
fig1, ax1 = plt.subplots()  # a figure with a single Axes
ax1.set_title('My Title')  # title above the plot
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# no plot.show() needed for labels to show up
```

## 2x2 Figure

axs2 is a _grid of axes_ objects, each containing _2 axis objects_.


```python
fig2, axs2 = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
axs2[0][1].set_title('Title 0,1')  # becomes a 2d array with [row][col]
axs2[1][1].set_xlabel(
    'My Independent Variable')  # axis labels are independent for each
axs2[0][0].set_ylabel('My Dependent Variable')  # setting y and x labels
```

## Single Dimension Figure Lists


```python
fig3, axs3 = plt.subplots(2, 1)
axs3[0].set_title('hi')  # axs3[0][0] would be an error!
# subplots(1, 2) would also give you a single dimension list (but look like a row vector)
# can also tuple-unwrap it on declaration!
```

## Sizing and layout

_figsize_ is width x height in _inches_.

_constrained layout_ makes legend, etc. auto-adjust better.


```python
plt.subplots(figsize=(2, 3))
plt.subplots(figsize=(2, 3), layout='constrained')
```

## Mosaic


```python
fig, axd = plt.subplot_mosaic([['upleft', 'right'], ['lowleft', 'right']],
                              layout='constrained')
axd['upleft'].set_title('upleft')
axd['lowleft'].set_title('lowleft')
axd['right'].set_title('right')
```

## Hide both axes


```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 2, 3])
ax.axis('off')
```

## Hide one axis


```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 2, 3])
ax.xaxis.set_visible(False)
```

## Remove border but leave axes


```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 2, 3])

# You can pick whichever ones you want to keep.
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
```

## Implicit API

NOTE: implicit and explicit can be mixed (eg. create with explicit then use implicit)


```python
# Creation
plt.figure(figsize=(6, 6))  # 6"x6" figure (no axes objects yet)
plt.subplot(
    1, 2,
    1)  # 1st axes object created w/ assumption of 1x2 grid (1-based index)
plt.subplot(
    1, 2,
    2)  # 2nd axes object created w/ same assumptions (and using next index)

# Plotting
plt.plot([1, 2, 3],
         [4, 5, 6])  # Last plot created automatically the one manipulated
plt.subplot(1, 2, 1)
plt.subplot(1, 2, 2)
plt.plot(
    [1, 2, 3],
    [7, 8, 9
    ])  # You can navigate the subplots again after you already used them before

# Creating another figure
plt.figure(figsize=(5, 5))
plt.subplot(2, 2,
            2)  # Does not automatically add all 4 (only each one you index)
plt.plot(
    [1, 2, 3],
    [4, 5, 6])  # Any kind of plot will work (eg. TF labs use imshow() a lot)
plt.subplot(2, 2, 1)  # Slides into the right spot based on the grid

# Labels and axes
plt.title("my title")  # Like the explicit API but without set_ in front
plt.xlabel("my x label")
plt.subplot(2, 2, 2)
plt.axis(
    'off'
)  # Some methods are exactly the same as explicit because never had set_ in front
```

## Implicit API - new cell


```python
plt.plot([1, 2, 3], [4, 5, 6])  # Starts a new figure because in new cell
```

## Layout Types


```python
fig, axs = plt.subplots(2, 2)
plt.title('default')
fig, axs = plt.subplots(2, 2, layout='constrained')
plt.title('constrained')
fig, axs = plt.subplots(2, 2, layout='tight')
plt.title('tight')

# Implicit API version
plt.tight_layout()
```

## Controlling where Graph Appears


```python
print('before graph 1')
plt.plot([1, 2, 3], [4, 5, 6])  # Graph 1 does not appear yet
print('after graph 1')

print('before graph 2')
plt.plot([1, 2, 3], [4, 5, 7])
plt.show(
)  # This ends up showing both graphs at the same time between these 2 prints
print('after graph 2')

print('before graph 3')
plt.plot([1, 2, 3],
         [4, 5, 8])  # Automatically shown at the very end of the cell execution
print('after graph 3')
```

## plot.show() Completes a Figure


```python
# These plot to the same figure
plt.plot([1, 2, 3], [4, 5, 6])
plt.plot([1, 2, 3], [4, 5, 7])
plt.show()

# This plots to a new figure because of previous 'show'
plt.plot([1, 2, 3], [4, 5, 8])
plt.show()

# A new figure again
fig, axis = plt.subplots()
axis.plot([1, 2, 3], [4, 5, 9])
plt.show()

# This has NO EFFECT because 'show' already completed the figure
axis.plot([1, 2, 3], [4, 5, 5])
plt.show()
pass
```
